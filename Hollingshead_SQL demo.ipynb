```python
# Import libraries
import pandas as pd
import pyodbc
```

```python
server = 'localhost'
port = '1433'
database = 'master'
username = 'sa'
password = 'YourStrong@Passw0rd'

query = "SELECT TOP 5 * FROM list_a"

def sqlserver_connect(query):
    conn_str = (
        'DRIVER={ODBC Driver 17 for SQL Server};'
        f'SERVER={server},{port};'
        f'DATABASE={database};'
        f'UID={username};'
        f'PWD={password}'
    )
    
    conn = pyodbc.connect(conn_str)
    
    conn_df = pd.read_sql(query, conn)
    
    conn.close()
    return conn_df

if __name__ == "__main__":
    sqlserver_connect(query)
```

```python
list_a_query = "SELECT TOP 10 * FROM list_a"
list_a_df = sqlserver_connect(list_a_query)
list_a_df.head()
```

```python
list_b_query = "SELECT TOP 10 * FROM list_b"
list_b_df = sqlserver_connect(list_b_query)
list_b_df.head()
```
